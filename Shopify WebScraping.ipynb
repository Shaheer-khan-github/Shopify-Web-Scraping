#### Import Packages

```python
# import dependencies
import requests
import json
import pandas as pd

# define the url
# url = 'https://helmboots.com/products.json'
url = 'https://partakefoods.com/products.json'
```

#### Json Data 

```python
# Get url data using requests
r = requests.get(url)
# Convert to json
json_data= r.json()
```

#### Assign the product key values to the variables

```python
# List of product list
product_list = []

# Loop on products key and assign it's values to the variables
for item in json_data['products']:
    id = item['id']
    title = item['title']
    handle = item['handle']
    body_html = item['body_html']
    created_at = item['created_at']
    updated_at = item['updated_at']
    vendor = item['vendor']
    product_type = item['product_type']

    # Nested keys of product key
    for variant in item['variants']:
        variant_id = variant['id']
        variant_title = variant['title']
        variant_price = variant['price']
        variant_sku = variant['sku']
        variant_position = variant['position']
        variant_grams = variant['grams']
        variant_compare_at_price = variant['compare_at_price']        
        variant_requires_shipping = variant['requires_shipping']
        variant_taxable = variant['taxable']
        variant_featured_image = variant['featured_image']
        variant_available = variant['available']
        variant_created_at = variant['created_at']
        variant_updated_at = variant['updated_at']
        variant_product_id = variant['product_id']
        variant_option1 = variant['option1']
        variant_option2 = variant['option2']
        variant_option3 = variant['option3']
    for image in item['images']:
        image_id = image['id']
        image_position = image['position']
        image_created_at = image['created_at']
        image_updated_at = image['updated_at']
        image_product_id = image['product_id']
        image_src = image['src']
        image_variant_ids = image['variant_ids']
        image_width = image['width']
        image_height = image['height']
    for option in item['options']:
        option_name = option['name']
        option_position = option['position']
        option_values = option['values']

        # product Dictionary of variables
        product = { 'id': id, 
                    'title': title, 
                    'handle': handle, 
                    'body_html': body_html,
                    'created_at': created_at, 
                    'updated_at': updated_at, 
                    'vendor': vendor, 
                    'product_type': product_type, 
                    'variant_id': variant_id, 
                    'variant_title': variant_title, 
                    'variant_price': variant_price, 
                    'variant_sku': variant_sku, 
                    'variant_position': variant_position, 
                    'variant_grams': variant_grams, 
                    'variant_compare_at_price': variant_compare_at_price, 
                    'variant_requires_shipping': variant_requires_shipping, 
                    'variant_taxable': variant_taxable, 
                    'variant_featured_image': variant_featured_image, 
                    'variant_available': variant_available, 
                    'variant_created_at': variant_created_at, 
                    'variant_updated_at': variant_updated_at, 
                    'variant_product_id': variant_product_id, 
                    'variant_option1': variant_option1, 
                    'variant_option2': variant_option2, 
                    'variant_option3': variant_option3, 
                    'image_id': image_id, 
                    'image_position': image_position, 
                    'image_created_at': image_created_at, 
                    'image_updated_at': image_updated_at, 
                    'image_product_id': image_product_id, 
                    'image_src': image_src, 
                    'image_variant_ids': image_variant_ids, 
                    'image_width': image_width, 
                    'image_height': image_height, 
                    'option_name': option_name, 
                    'option_position': option_position, 
                    'option_values': option_values
                }
        # Append the product dictionary to the product list
        product_list.append(product)
```

#### Create Dataframe

```python
# Product list to dataframe
df = pd.DataFrame(product_list)
df.head()
```
